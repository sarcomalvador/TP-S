Este conjunto de datos procede del Instituto Nacional de Diabetes y Enfermedades Digestivas y Renales. La selección de estas instancias de una base de datos más amplia está sujeta a varias restricciones como por ejemplo se trata de pacientes mujeres.

- Pregnancies: Número de embarazos

- Insulin: Insulina sérica de 2 horas (mu U/ml)
- BMI: Índice de masa corporal (peso en kg/(altura en m)^2)
- DiabetesPedigreeFunction: Función del pedigrí de la diabetes
- Age: Edad (años)
- Outcome: Variable de clase (0 o 1) 1 para personas diabéticas, 0 para las que no lo son.

##### **Objetivo 11**

Hacer un boxplot de la distribución de la "Glucosa" segun el tipo de paciente de la variable "Outcome". Es decir que hay que visualizar un boxplot de la variable "Glucosa" por cada categoria de la variable "Outcome".

Que podemos observar?

```python
g = sns.catplot(data=diabetes_set,
                x='Outcome',y='Glucose', 
                kind="box", height = 5 ,aspect=3,palette = "muted")
g.despine(left=True)
g.set_xticklabels(rotation=90)
plt.xlabel("Outcome",fontsize=32)
plt.ylabel("Glucosa",fontsize=32)
plt.title('distribución de la "Glucosa" segun el tipo de paciente de la variable "Outcome"',size = 42)
plt.show()
```

##### **Objetivo 13**
Crear una nueva variable de rangos de edades de tipo texto/ string dividiendolo de la siguiente manera:

- Primer rango de edad menor a 30 años
- Segundo rango de edad entre 30 y 40 años.
- Tercer rango de edad mayor a 40 años

```python
#df.loc() filtra por labels (y por eso se pueden usar 'nombres' de columna/filas)
#fue necesario crear la columna rango, creo que no asi definir como int la columna "age"
diabetes_set=diabetes_set.assign(RANGO="")
diabetes_set['Age'].astype(int)
diabetes_set.loc[(diabetes_set['Age'] > 0) & (diabetes_set['Age'] <= 30),'RANGO'] = 'Primer'
diabetes_set.loc[(diabetes_set['Age'] > 30) & (diabetes_set['Age'] <= 40),'RANGO'] = 'Segundo'
diabetes_set.loc[diabetes_set['Age'] > 40, 'RANGO'] = 'Tercero'
diabetes_set.head()
```

##### **Objetivo 14**
Hacer un count plot que cuente la cantidad de pacientes diabéticos y no diabéticos por rango etario.

Podemos extraer alguna conclusión al respecto? Cuál?

```python
sns.countplot(data=diabetes_set, x="Outcome", hue="RANGO")
#la conclusion que se obtiene es la gran cantidad de gente menor de 30 que no tiene diabetes
#no solo comparado con su mismo rango, sino que triplica a todos los demas
#dando una gran probabilidad de no tener diabetes si se tiene menos de 30 años
```
